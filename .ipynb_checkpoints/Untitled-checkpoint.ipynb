```python
plt.figure(figsize=(7,5))
a = sns.distplot(df[df['gender']=='Female']['cust_score'], hist= False, label = 'Female')
a = sns.distplot(df[df['gender']=='Male']['cust_score'], hist= False, label = 'Male')

a.set(xlabel = 'Gender', ylabel = 'Density')
plt.legend()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
lm = ols('cust_score ~ gender', data=df).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

```python
lm = ols('cust_score ~ favorite_time', data= df).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

```python
sns.boxplot(x= 'favorite_time', y='cust_score', data= df)
```

```python
plt.figure(figsize=(7,5))
a = sns.distplot(df[df['favorite_time']=='Weekday']['cust_score'], hist= False, label = 'Weekday')
a = sns.distplot(df[df['favorite_time']=='Weekend']['cust_score'], hist= False, label = 'Weekend')

a.set(xlabel = 'Favorite Time for Shopping', ylabel = 'Density')
plt.legend()
```
